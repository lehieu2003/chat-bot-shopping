# HTML Tag Cleaning
The material column contains HTML tags that need to be removed for better data analysis.

```python
# Install BeautifulSoup if not already installed
# !pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
import re

# Function to clean HTML tags
def clean_html(html_text):
    if pd.isna(html_text):
        return ""
    # Use BeautifulSoup to remove HTML tags
    soup = BeautifulSoup(html_text, 'html.parser')
    clean_text = soup.get_text(separator=' ')
    # Remove extra whitespace
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    return clean_text

# Apply cleaning to material columns
fashion_ao['material_clean'] = fashion_ao['material'].apply(clean_html)
fashion_quan['material_clean'] = fashion_quan['material'].apply(clean_html)
```

```python
# Check cleaned material columns
print("Fashion Áo - Original vs Cleaned:")
for i in range(2):
    print(f"\nOriginal: {fashion_ao['material'].iloc[i]}")
    print(f"Cleaned: {fashion_ao['material_clean'].iloc[i]}")

print("\n\nFashion Quần - Original vs Cleaned:")
for i in range(2):
    print(f"\nOriginal: {fashion_quan['material'].iloc[i]}")
    print(f"Cleaned: {fashion_quan['material_clean'].iloc[i]}")
```

```python
# Option 1: Replace the original material columns
fashion_ao['material'] = fashion_ao['material_clean']
fashion_quan['material'] = fashion_quan['material_clean']

# Drop the temporary columns
fashion_ao.drop(columns=['material_clean'], inplace=True)
fashion_quan.drop(columns=['material_clean'], inplace=True)
```

```python
# Option 2: Save cleaned dataframes to new CSV files
fashion_ao.to_csv('D:/chat-bot/exports/fashion_áo_cleaned.csv', index=False)
fashion_quan.to_csv('D:/chat-bot/exports/fashion_quần_jean_cleaned.csv', index=False)

print("Cleaned data saved to CSV files.")
```
